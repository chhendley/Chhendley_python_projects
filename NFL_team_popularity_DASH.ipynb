```python
import pandas as pd
import numpy as np
from matplotlib import animation
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import seaborn as sns
%matplotlib inline
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objects as go
```

```python
pop_2020 = pd.read_csv("https://github.com/chhendley/chhendley_python_projects/raw/main/US%20team%20popularity%202020.csv")
pop_2021 = pd.read_csv('https://github.com/chhendley/chhendley_python_projects/raw/main/US%20team%20popularity%202021.csv')
pop_2022 = pd.read_csv('https://github.com/chhendley/chhendley_python_projects/raw/main/US%20team%20popularity%202022.csv')

```

```python
total_pop = pd.concat([pop_2020, pop_2021, pop_2022], axis=0)
display(total_pop.columns)
```

```python
total_pop.rename(columns={'Arizona Cardinals': ' Arizona Cardinals', 'date': 'Date'}, inplace=True)
total_pop = total_pop.reindex(sorted(total_pop.columns), axis=1)
#total_pop.set_index(total_pop['Date'], inplace=True)
total_pop.drop(total_pop.columns[[-1]],axis =1, inplace=True)
```

```python
# Convert 'date' column to datetime format
total_pop['Date'] = pd.to_datetime(total_pop['Date'])
team_cols = total_pop.loc[:, ' Arizona Cardinals':' Washington Commanders']
# Set index to Date column
#total_pop.set_index('Date', inplace=True)
#display(total_pop.columns)
```

```python
import dash
from dash import dcc, html
import pandas as pd
import plotly.graph_objs as go

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Checklist(
        id='team-checkboxes',
        options=[{'label': col, 'value': col} for col in total_pop.columns[1:]],  # Exclude the 'Date' column
        value=[],  # Set default selected teams here
        labelStyle={'display': 'inline-block'}
    ),
    dcc.Graph(id='line-chart')
])

@app.callback(
    dash.dependencies.Output('line-chart', 'figure'),
    [dash.dependencies.Input('team-checkboxes', 'value')]
)
def update_line_chart(selected_teams):
    data = []
    for team in selected_teams:
        data.append(go.Scatter(
            x=total_pop['Date'],
            y=team_cols,
            mode='lines+markers',
            name=team
        ))

    layout = go.Layout(title='Football Team Performance Over Time',
                       xaxis=dict(title='Date'),
                       yaxis=dict(title='Value'))

    return {'data': data, 'layout': layout}

if __name__ == '__main__':
    app.run_server(debug=True)
```
